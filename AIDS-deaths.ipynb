Looking for trends

Finding trends on this data is important because it can indicate which countries have successful AIDS prevention programs, and which need more multilateral intervention to save lives. 

A widely used test for monotonic trends is the Mann-Kendall (MK) test whose purpose is to statistically assess if there is a monotonic upward or downward trend of the data over time. A monotonic upward (downward) trend means that the variable consistently increases (decreases) through time, but the trend may or may not be linear. 

source: http://vsp.pnnl.gov/help/Vsample/Design_Trend_Mann_Kendall.htm

```python
import numpy as np  
from scipy.stats import norm, mstats

#From http://michaelpaulschramm.com/simple-time-series-trend-analysis/
def mk_test(x, alpha = 0.05):  
    """   
    Input:
        x:   a vector of data
        alpha: significance level (0.05 default)

    Output:
        trend: tells the trend (increasing, decreasing or no trend)
        h: True (if trend is present) or False (if trend is absence)
        p: p value of the significance test
        z: normalized test statistics 

    Examples
    --------
      >>> x = np.random.rand(100)
      >>> trend,h,p,z = mk_test(x,0.05) 
    """
    n = len(x)

    # calculate S 
    s = 0
    for k in range(n-1):
        for j in range(k+1,n):
            s += np.sign(x[j] - x[k])

    # calculate the unique data
    unique_x = np.unique(x)
    g = len(unique_x)

    # calculate the var(s)
    if n == g: # there is no tie
        var_s = (n*(n-1)*(2*n+5))/18
    else: # there are some ties in data
        tp = np.zeros(unique_x.shape)
        for i in range(len(unique_x)):
            tp[i] = sum(unique_x[i] == x)
        var_s = (n*(n-1)*(2*n+5) + np.sum(tp*(tp-1)*(2*tp+5)))/18

    if s>0:
        z = (s - 1)/np.sqrt(var_s)
    elif s == 0:
            z = 0
    elif s<0:
        z = (s + 1)/np.sqrt(var_s)

    # calculate the p_value
    p = 2*(1-norm.cdf(abs(z))) # two tail test
    h = abs(z) > norm.ppf(1-alpha/2) 

    if (z<0) and h:
        trend = 'decreasing'
    elif (z>0) and h:
        trend = 'increasing'
    else:
        trend = 'no trend'

    return trend, h, p, z
```

The first test will be with South Africa, where according to the visualization there is an upward trend until 2006 and then a downward trend

```python
data2.T["South Africa"]
```

```python
mk_test(list(data2.T["South Africa"]['1990':'2005']))
```

```python
mk_test(list(data2.T["South Africa"]['2006':'2010']))
```

According to the test, the upward trend from 1990 to 2005 is statistically significant (p = 8.34e-08) but it is not possible to say that there is a downward trend from 2006 to 2010 because the p-value is not smaller than 0.05  (p=0.057), probably because the number of values is small

A second test with Burkina Faso, where there should be an upward trend from 1990 to 2001, and then a downward trend until 2011

```python
data2.T["Burkina Faso"]
```

```python
trend, _, pval, _ = mk_test(list(data2.T["Burkina Faso"]['1990':'2001']))
trend, pval
```

```python
trend, _, pval, _ = mk_test(list(data2.T["Burkina Faso"]['2001':'2011']))
trend, pval
```

The trends are statistically significant in this case

The idea is now to explore the entire dataset and find which countries have downward trends on the last decade, 

```python
def mk_test_trend(x, alpha = 0.05):
    trend, _, pval, _ = mk_test(x, alpha)
    return trend
```

```python
data2.T.apply(lambda x: mk_test_trend(list(x['2001':'2011'])))
```

```python
successful_countries = data2.T.apply(lambda x: mk_test_trend(list(x['2001':'2011']))) == "decreasing"
```

```python
successful_countries
```

Visualizing countries with successful prevention programs:

```python
plt.figure()
data2[successful_countries].T.plot()
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=6)
```

Visualizing countries where AIDS is worsening:

```python
worsening_countries = data2.T.apply(lambda x: mk_test_trend(list(x['2001':'2011']))) == "increasing"
```

```python
plt.figure()
data2[worsening_countries].T.plot()
ax = plt.subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 0.9])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=6)
```

With this information a multilateral organization like the World Health Organization could focus their resources to prevention programs on countries were the situation is worsening, or study the efforts of countries where the number of AIDS deaths is decreasing
